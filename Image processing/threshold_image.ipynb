```python
ret, thresh1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
ret, thresh2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
ret, thresh3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
ret, thresh4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)
```

```python
img = cv.imread('images/crossword.jpg', 0)
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv.threshold(img, 160, 255, cv.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 9, 16)
show_pic(th2)
```

```python
blended = cv.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
